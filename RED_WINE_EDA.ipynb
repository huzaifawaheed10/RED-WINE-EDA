## TASK 5 : EDA

```python
data=df[df['quality'] >3]

sns.boxplot(data=data,x='quality')
plt.grid(False)
```

```python
sns.histplot(data=data,x='quality')
plt.title('Wine quality distribution')
plt.grid(False)
plt.show()
```

```python
for column in df.columns:
    sns.histplot(df[column],kde=True)
plt.grid(False)
```

```python
sns.histplot(df['alcohol'])
plt.grid(False)
```

```python
sns.pairplot(df)
```

```python
sns.catplot(data=df,x='quality',y='alcohol',kind='box')
plt.grid(False)
```

```python
sns.scatterplot(data=df,y='pH',x='alcohol',hue='quality')
plt.grid(False)
```
